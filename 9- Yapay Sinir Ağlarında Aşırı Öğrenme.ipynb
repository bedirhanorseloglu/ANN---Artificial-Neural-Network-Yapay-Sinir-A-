# Aşırı Öğrenme 



- Yapay Sinir Ağlarında Aşırı Öğrenmeyi durdurmak için belirli yöntemler kullanacağız:
    - Erken Durdurma (Early Stopping)
    - L1 ve L2 Düzenlileştirme
    - Sönümleme (Dropout)
    - Veri Artırımı (Data Augmentation)

# L1 ve L2 Düzenlileştirme

- Amaç hata fonksiyonunu minimize eden katsayıları (ağırlıkları), bu katsayılara bir ceza uygulayarak bulmaktır.



# Sönümleme (Dropout)

- Sinir ağı içerisinde yer alan nöronların kullanıcı tarafından belirlenen bir oranda rastgele olarak söndürülmesiyle aşırı öğrenmeyi önleyerek performansın artırılmasını amaçlamaktadır.
- Bu şekilde bazı nöronları söndürdüğümüzde diğer nöronların katsayıları daha da önem kazanarak nöron kendini daha iyi ifade edebilir. Bu şekilde işin içine rastgelelik de katarak daha genellenebilir bir model elde etmiş oluruz.




# Veri Artırımı (Data Augmentation)

- Veri artırımı, veri setinde bulunan verilere çeşitli özellikler ekleyerel veri miktarını artırmayı sağlayan bir tekniktir.
- Örneğin bir resmi dikey , yatay , eğik , çevirme , zoom yapma gibi farklı biçimlerde şekillendirmek
