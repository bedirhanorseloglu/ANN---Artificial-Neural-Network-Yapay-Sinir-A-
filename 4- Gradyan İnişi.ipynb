- Bir önceki bölümde fark ettik ki ağırlıkları değiştirdiğimizde sonuç değişiyor. Bu ağırlıkları belirli bir kurala, belirli bir yönteme göre gerçekleştirebilirsek bu durumda optimum gelinmesi gereken noktaya gelebileceğimizi, optimum ağırlık değerlerini bulabileceğimizi söylüyoruz.

- Gradient Descent (Gradyan İniş) bunu gerçekleştirmek üzere ortaya çıkan bir optimizasyon yöntemidir. 





- Solda çizdiğimiz çizginin hata oranı(Cost , Lost = MSE) olduğunu düşünelim. Bunun haricinde iki eksenimiz var x ve y. Hedefimiz x ve y çifti bir araya geldiğinde oluşturulacak olan değerin hata oranı ekseninde 0'a yaklaşması. Burada x ve y ifadelerini w(ağırlıklar) olarak düşünecek olursak çeşitli ağırlık değişimlerinde çeşitli hata oranları alırız.





- **Gradient Descent Nasıl Çalışır?**
    - Gradyanın(türev sonucu elde edeceğimiz değerlerin) negatifi olarak tanımlanan en dik iniş yönünde iteratif olarak parametre değerlerini güncelleyerek ilgili fonksiyonun minimum değerini verebilecek parametreleri bulur.

- Gradient Descent, Cost(hata oranı) fonksiyonunu minimize edebilecek parametreleri bulmak için kullanılır.

- **Gradien Descent kademeli şekilde gitmesi gereken en doğru yolu nasıl belirliyor?**
    - Parametre değerlerini iteratif olarak cost'u azaltacak şekilde değiştirerek yapar.
    - Bir fonksiyonun o noktadaki türevi o fonksiyonun maksimum artış yönünü verir. Bundan dolayı Gradient Descent bu artış yönünün tersine gider.
    - Cost fonskiyonunun bir noktadaki(w) türevi cost fonksiyonunun maksimum artış yönünü verir. Bunun tersine gidildiğinde maksimum azaldığı yöne doğru gidilmiş olur. 
    
    - Yani w'ya göre kısmi türev alındığında cost'u en fazla arttıracak yön bilgisi bulunur. Bu yönün tersine gidildiğinde cost'un en hızlı azalacağı tarafa doğru gidilmiş olur(ilk görseldeki "-" ifadesi). Belirli bir gidiş hızında (learning rate - α ) ve belirli bir iterasyon sayısınca bu işlem tekrar edilir.
