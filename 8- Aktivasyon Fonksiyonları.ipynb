- Aktivasyon fonksiyonları, karmaşık yapıdaki veriler içierisinden yapay sinir ağlarının öğrenmesini sağlayabilmek adına kullanılan fonksiyonlara denir.

- **Sigmoid Fonksiyonu:** 0-1 arasında değer üretmeyi sağlar. Bu sayede 0.5 değerini eşik değer olarak belirleyerek bundan küçükler 0, bundan büyükler 1 sınıfna aittir gibi yorum yapmamızı sağlar.  Genelde ikili sınıflandırma problemleri için kullanılır. Uç noktalarda değerler girildiğinde 0'a yakınsar. Bu duruma <b style="color:red">Kaybolan Gradyan Problemi</b>  denir. Bu geriye yayılım sırasında öğrenmenin durmasına sebep olur. Bu sebeğle çok katmanlı ağlarda özellikle gizli katmanlarda kullanılması tercih edilmez.
 
 - **Hiperbolik Tanjant Fonksiyonu:** Sigmoid'e oldukça benzerdir. -1 ve +1 arasında değer üretmeyi sağlar. Doğrusal olmayan bir fonksiyondur. Doğrusal olmayan fonksiyonların temsil kabiliyeti daha yüksektir. Uç noktalarda değerler girildiğinde 0'a yakınsar. Bu duruma <b style="color:red">Kaybolan Gradyan Problemi</b>  denir. Bu geriye yayılım sırasında öğrenmenin durmasına sebep olur. Bu sebeğle çok katmanlı ağlarda özellikle gizli katmanlarda kullanılması tercih edilmez.


 - **ReLU Fonksiyonu:**  Diğer bir adıyla "Doğrultulmış Lineer Birim", negatif değerler için 0 değerini alırken pozitif girdiler için çeşitli değerler alabilmektedir. Genellikle gizli katmanlardaki nöronlarda kullanılır. Nasıl ki insan sinir hücreleri kullanılmadığında ya da gerekli işlevi yerine getiremediğinde ölüyorsa **relu fonksiyonu**  da yapay sinir ağı açısından da gelen bilgi 0'dan küçükse bunu aktif etmeyecektir ve o nöron 0 değerini alarak etkisiz hale gelir.
