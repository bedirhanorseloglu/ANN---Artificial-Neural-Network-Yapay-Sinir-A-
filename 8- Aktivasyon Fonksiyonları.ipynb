- Aktivasyon fonksiyonlar, karmak yapdaki veriler iierisinden yapay sinir alarnn renmesini salayabilmek adna kullanlan fonksiyonlara denir.

- **Sigmoid Fonksiyonu:** 0-1 arasnda deer retmeyi salar. Bu sayede 0.5 deerini eik deer olarak belirleyerek bundan kkler 0, bundan bykler 1 snfna aittir gibi yorum yapmamz salar.  Genelde ikili snflandrma problemleri iin kullanlr. U noktalarda deerler girildiinde 0'a yaknsar. Bu duruma <b style="color:red">Kaybolan Gradyan Problemi</b>  denir. Bu geriye yaylm srasnda renmenin durmasna sebep olur. Bu sebele ok katmanl alarda zellikle gizli katmanlarda kullanlmas tercih edilmez.
 
 - **Hiperbolik Tanjant Fonksiyonu:** Sigmoid'e olduka benzerdir. -1 ve +1 arasnda deer retmeyi salar. Dorusal olmayan bir fonksiyondur. Dorusal olmayan fonksiyonlarn temsil kabiliyeti daha yksektir. U noktalarda deerler girildiinde 0'a yaknsar. Bu duruma <b style="color:red">Kaybolan Gradyan Problemi</b>  denir. Bu geriye yaylm srasnda renmenin durmasna sebep olur. Bu sebele ok katmanl alarda zellikle gizli katmanlarda kullanlmas tercih edilmez.


 - **ReLU Fonksiyonu:**  Dier bir adyla "Dorultulm Lineer Birim", negatif deerler iin 0 deerini alrken pozitif girdiler iin eitli deerler alabilmektedir. Genellikle gizli katmanlardaki nronlarda kullanlr. Nasl ki insan sinir hcreleri kullanlmadnda ya da gerekli ilevi yerine getiremediinde lyorsa **relu fonksiyonu**  da yapay sinir a asndan da gelen bilgi 0'dan kkse bunu aktif etmeyecektir ve o nron 0 deerini alarak etkisiz hale gelir.
