# Yapay Sinir Ağı

- 1957 yılında Frank Rosenblatt tarafından tanımlanmıştır. 5 tonluk, bir oda büyüklüğündeki bir bilgisayar olan "IBM704" bir dizi delikli kart ile beselenerek 50 deneme sonrasında birbirinden farklı işaretlenmelere sahip olan kartları birbirinden ayırt edebilir bir hale gelmiştir. Bu durum ile ilgili Rosenblatt şöyle bir ifadede bulunmuştur: "Orijinal bir fikre sahip olabilecek ilk makine".

- YSA'daki amaç makinelere karar verebilme, kendi kendine öğrenebilme ve değerlendirebilme gibi yetkinlikleri kazandırarak insan beynini modellemeye çalışmaktır.

- Katmanlar ham halde gelen veriyi işleyerek daha kullanışlı çıkarımlar yapar ve ilgili problemin çözümünde yardımcı olabilecek daha anlamlı bilgilerin çıkmasına katkı sağlar.

- Giriş katmanı, sinir ağına veriyi gönderdiğimiz katmandır. 

- Gizli katman, giriş ile çıkış arasındaki katmandır/katmanlardır. Öğrenme işlemi bu katmanda gerçekleşir. 

- Çıktı katmanı, giriş değerlerine göre öğrenilmesi istenen çıkış verileri buraya gelir.





# Derin Sinir Ağları

- Franco Chollet (Keras Kütüphanesi Kurucusu), derin sinir ağları için şunu der: "Derin sinir ağları bilgi damıtma hattı gibi düşünebilir. Ham veri ilerledikçe tekraren dönüşüme uğrar ve gereksiz bilgiler atılırken kullanışlı veriye yaklaşılır. Yani veri arıtılır."

- Andrew Ng der ki "Beyindeki tek bir nöron bugün bile hala anlamlandıramadığımız inanılmaz derecede karmaşık bir makinedir. Bir sinir ağındaki tek bir "nöron", biyolojik bir nöronun karmaşıklığının çok küçük bir bölümünü yakalayan inanılmaz derecede basit bir matematiksel işlevdir"
